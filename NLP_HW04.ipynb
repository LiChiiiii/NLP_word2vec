### 讀所有檔案 -> 簡體轉繁體存進 dataset.txt

```python
from os import listdir
from os.path import isdir
from opencc import OpenCC

#簡體轉繁體
def simplified_to_traditional(path):
    cc = OpenCC('s2twp')
    traditional = open("dataset.txt", "a", encoding = "utf-8")
    with open(path, "r", encoding = "utf-8") as simplified:
        for s in simplified:
            traditional.write(cc.convert(s))
    traditional.close()
#讀取資料
def load_file(directory):
    file = list()
    for filename in listdir(directory):
        path = directory + filename
        print(path)
        simplified_to_traditional(path)
#讀取資料夾
def load_directory(directory):
    # enumerate folders, on per class
    for subdir in listdir(directory):
        path = directory + subdir + '/'
        # skip any files that might be in the dir
        if not isdir(path):
            continue
        load_file(path)
    return "success"

###main
load_directory('wiki_zh/')
```

### word2vec訓練

```python
from gensim.models import word2vec

def train():
    print("訓練中...")
    sentence = word2vec.Text8Corpus("segmentation.txt")
    model = word2vec.Word2Vec(sentence, vector_size=400,epochs=5, workers = 4, sg = 1)
    model.wv.save_word2vec_format(u"wiki300.model.bin", binary = True)
    print("model 儲存完畢")

train()
```
